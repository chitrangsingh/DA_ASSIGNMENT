```python
wage_by_age = fifa.groupby('age')['wage_eur'].mean()
wage_by_age.plot(kind='line', marker='o', color='blue', figsize=(10, 6))
plt.title('Average Wage by Age')
plt.xlabel('Age')
plt.ylabel('Average Wage (EUR)')
plt.grid(True)
plt.show()
```

```python
shooting_by_age = fifa.groupby('age')['shooting'].mean()
shooting_by_age.plot(kind='bar', color='lightgreen')
plt.title('Average Shooting Ability by Age')
plt.xlabel('Age')
plt.ylabel('Average Shooting')
plt.show()
```

```python
top_5_nations = fifa['nationality'].value_counts().head(5)
top_5_nations.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8), colormap='Set3')
plt.title('Top 5 Nationalities in FIFA Dataset')
plt.show()
```
